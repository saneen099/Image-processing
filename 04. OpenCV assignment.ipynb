### Import necessary Libraries

```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt 
```

### Read 'image1.jpg' with opencv

```python
img = cv2.imread('image1.jpg')
```

```python
print(img)
```

### Plot the image with matplotlib

```python
plt.imshow(img)
```

### Convert the image to RGB

```python
new_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(new_img)
```

### Check the shape of the image

```python
new_img.shape
```

### Resize the image to (400,300)

```python
img = cv2.resize(new_img,(400,300))
```

```python
plt.imshow(img)
```

### Resize the image by ratio. Use 0.5 height and width

```python
h_img = 0.5
w_img = 0.5
```

```python
img3 = cv2.resize(new_img,(0,0),img,h_img,w_img)
plt.imshow(img3)
```

### Flip the image along central x axis

```python
img3 = cv2.flip(img3,0)
plt.imshow(img3)
```

### Flip the image along central y axis 

```python
img3 = cv2.flip(img3,1)
plt.imshow(img3)
```
